```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
from train_EuroSAT import BS_VALUES as values

r = {}
for bs in values:
    with open(f"results-{bs}.dat", "rb") as fh:
        r[bs] = pickle.load(fh)
```

```python
labels = [str(bs) for bs in values]
times = [r[bs]['time'] for bs in values]
accs = [100*r[bs]['best_accuracy'] for bs in values]

(fig, ax) = plt.subplots(2, 1)
ax[0].set_ylabel("training time")
ax[0].bar(labels, times)
ax[1].set_ylabel("best accuracy")
ax[1].bar(labels, accs)
#ax[1].set_ylim(85, 100)
plt.show()
```

```python
for (i,bs) in enumerate(values):
    d = np.array(r[bs]['epochs'])
    plt.figure()
    plt.title(f"batch size = {bs}")
    p1, = plt.plot(d[:,0], "r", label="loss")
    plt.twinx()
    p2, = plt.plot(100*d[:,1], "g", label="validation accuracy")
    plt.legend(handles=[p1,p2])
plt.show()
```
